```python
#Preparación de Datos para Apriori
rawapri=raw_fmerge.copy()
rawapri=rawapri.set_index("ID").drop(columns=["Idade","Renda","Regiao","Genero"])
rawapri
```

```python
#Enconding
def encode(x):
  if x=="T":
    return 1
  if x=="F":
    return 0

rawapri=rawapri.applymap(encode)
rawapri
```

```python
list(rawapri.columns)
porcentajes=[]
for column in list(rawapri.columns):
  porcentajes.append([column,(rawapri[column].sum()/13730)*100])


porcentajes_df=pd.DataFrame(porcentajes,columns=[column,"Porcentaje"])
porcentajes_df.round(2).sort_values(by="Porcentaje", ascending=False)
```

```python
#Importando Librería Apriori y Association Rules
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
#Apriori
frequent_itemsets=apriori(rawapri,min_support=0.10,use_colnames = True)
frequent_itemsets
```

```python
#Association Rules
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.8)
rules.sort_values('confidence', ascending = False, inplace = True)
rules
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
reglas=rules[ (rules['antecedent_len'] <= 5 )]
reglas
```
